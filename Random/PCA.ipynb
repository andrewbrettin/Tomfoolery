## Generate random data

```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt

from scipy.linalg import svd
from scipy.linalg import norm

rng = np.random.default_rng(0)
```

```python
mu = (5, 0)
A = np.array([
    [1, 2],
    [2, -5]
])
Sigma = np.transpose(A) @ A
X = rng.multivariate_normal(mu, Sigma, size=1000)
```

```python
fig, ax = plt.subplots()
ax.scatter(X[:,0], X[:,1], s=5, alpha=0.7)
ax.set(title='Raw data')
```

## Preprocessing

```python
# Need to shift data to have zero empirical mean
X = X - np.mean(X, axis=0)
```

```python
fig, ax = plt.subplots()
ax.scatter(X[:,0], X[:,1], s=5, alpha=0.7)
ax.set(title='Mean removed data')
```

## Compute SVD

```python
U, s, Vh = svd(X)
V = np.transpose(Vh)
```

```python
V[:,0]
```

```python
fig, ax = plt.subplots()
ax.scatter(X[:,0], X[:,1], s=5, alpha=0.7)
ax.arrow(0, 0, *V[0,:])
ax.arrow(0, 0, *V[1,:])
```
