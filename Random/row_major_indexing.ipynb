The purpose of this notebook is to see the importance of row-major indexing to performance in python.

```python
import numpy as np
from time import time
```

```python
arr = np.random.rand(10000, 10000)
```

```python
s = 0
t1 = time()
for i in range(1000):
    for j in range(1000):
        arr[i,j] #
t2 = time()
print(t2-t1)
```

```python
s = 0
t1 = time()
for j in range(1000):
    for i in range(1000):
        arr[i,j]
t2 = time()
print(t2-t1)
```

Doesn't look like there's any practical difference, though it does seem consistently slightly faster to run through the loop in column major order? Weird.
