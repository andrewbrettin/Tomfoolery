## Introduction

Want to find $\min f(x)$, where $f: \mathbb{R}^n \to \mathbb{R}$ is a $\mathcal{C}^1$ function.

**Gradient Descent**: $x_{k+1} = x_k - t_k \nabla f (x_k)$ Traverse direction of gradient.

**Newton's method**: $x_{k+1} = x_k - t_k (\nabla^2 f (x_k))^{-1} \nabla f(x_k)$ where $\nabla^2 f (x_k)$ is the Hessian of $f$.
