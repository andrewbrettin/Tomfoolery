```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt

from scipy.interpolate import CubicHermiteSpline
```

```python
xs = np.linspace(0,10,1001)
ys = (xs - 5) * (xs > 5)

plt.plot(xs, ys)
plt.title('Linear piecewise function')
```

```python
x1 = 3; x2 = 7
y1 = 0; y2 = 2
dydx1 = 0; dydx2 = 1

f = CubicHermiteSpline([3, 7], [0, 2], [0, 1])
plt.plot(xs, f(xs))
plt.plot([x1, x2], [y1, y2], ls = '', marker='o')

plt.title('Hermite interpolating polynomial')
```

```python
cond1 = (3 < xs) * (xs < 7)

zs = (cond1 * f(xs)) + (ys * (~cond1))
plt.plot(xs, zs)
plt.plot([x1, x2], [y1, y2], ls = '', marker='o')
plt.title('Smoothened corner using hermite interpolating polynomial')
```
