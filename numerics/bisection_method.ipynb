```julia
f(x) = (x - pi)^3
```

```julia
a = 0
b = 5
a_n = a
b_n = b
```

```julia
δ = 1e-8
n = log2((b - a)/δ) - 1
x_n = 0
for i in 1:n
    x_n = (a_n + b_n)/2
    
    if f(a_n)*f(x_n) < 0
        b_n = x_n
    elseif f(x_n)*f(b_n) < 0
        a_n = x_n
    end
        
end
x_n
# print("Root is:", x_n)
```
