```python
import numpy as np
import xarray as xr
import pandas as pd
import matplotlib.pyplot as plt
rng = np.random.default_rng(0)
```

```python
data = rng.random((60,3))
multiindex = pd.MultiIndex.from_product(
    [np.arange(10), list('ABCDEF')],
    names=['xt', 'yt']
)
df = pd.DataFrame(
    data, 
    index=multiindex,
    columns=['red', 'green', 'blue']
)
df.head()
```

```python
df = df[(df['red'] > 0.3) & (df['green'] > 0.3) & (df['blue'] > 0.3)]
```
