```python
import numpy as np
import pandas as pd
import xarray as xr
import matplotlib.pyplot as plt
```

```python
# Load and clean dataset
df = pd.read_hdf('/Users/andrewbrettin/Tomfoolery/data/cm2p6_df.h5')
df = df.dropna(axis=0)
```

```python
# Clustering
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

scaler = StandardScaler()
X = scaler.fit_transform(df[['std', 'skew', 'kurtosis']])
kmeans = KMeans(random_state=0).fit(X)
clusters = kmeans.predict(X)
df['cluster'] = clusters
```

```python
# Unsorted by standard deviation
grouped = df.groupby('cluster')
grouped['std'].mean()
```

```python
grouped['std'].mean().sort_values()
```

```python
d = dict(zip(grouped['std'].mean().sort_values().index, grouped['std'].mean().index))
d
```

```python
# In summary
grouped = df.groupby('Cluster')
d = dict(zip(grouped['std'].mean().sort_values().index, grouped['std'].mean().index))
df = df.replace({'cluster': d})
```
