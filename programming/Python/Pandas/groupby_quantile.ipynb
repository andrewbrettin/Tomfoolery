Here we explore how to use qcut to groupby a quantitative variable.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/andrewbrettin/Tomfoolery/data/battery_park_monthly_TG.csv')
df.head()
```

```python
# Break up MSL into bins with 4 quantiles
qcut = pd.qcut(df['MSL (ft)'], q=4)

# Create categorical variable based on quantile intervals
df['MSL quartiles'] = qcut
```

```python
# Apply groupby operations
gb = df.groupby('MSL quartiles')
gb['MHW (ft)'].mean()
```
