```python
import numpy as np
import pandas as pd
import xarray as xr
import sklearn
import matplotlib.pyplot as plt

import cftime
```

```python
# Load data
sea_level = pd.read_csv('~/Tomfoolery/data/battery_park_monthly_TG.csv')

def string_to_datetime(datestring):
    year = int(datestring[0:4])
    month = int(datestring[5:7])
    day = int(datestring[8:10])
    return cftime.datetime(year, month, day)

sea_level['Date'] = sea_level['Date'].apply(string_to_datetime)

# Drop nans
sea_level = sea_level['MSL (ft)'].dropna(axis=0)
```

```python
from sklearn import linear_model

# Select linear model
model = linear_model.LinearRegression()
```

```python
# Train model
X = np.array(sea_level.index).reshape(-1, 1)
Y = sea_level
model.fit(X, Y)
```

```python
# Make a prediction
pred = model.predict(X)
```
