```python
import numpy as np
import xarray as xr
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(
    np.random.random((1000, 3)),
    columns=['depth', 'A', 'B']
)
```

```python
df_groups = []
num_bins=10
quantiles = np.linspace(0,1,num_bins, endpoint=False)
quantile_spacing = 1/num_bins

for q in quantiles:
    next_q = q + quantile_spacing
    df_bin = df[
        (df['depth'].quantile(q) <= df['depth']) &
        (df['depth'] < df['depth'].quantile(next_q))
    ]
    df_groups.append(df_bin)
```
